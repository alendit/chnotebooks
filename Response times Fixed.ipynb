Response times
---------------

This is an analysis of the response times of the OLAP queries from the CHbenCHmark run. 

The OLAP query response time is dependent on the size of the database tables it is run against. In CHbenCHmark the size of the tables changes as the OLTP queries are executed. To compensate we will normalize the data using the model from Schreitz 2012.

The response times of OLAP queries grows linearly with the size of the tables which in turn grows linearly with the number of the OLTP queries. Using OLS linear model, we will find the dependecy between OLAP query runtime and number of executed OLTP queries and thus the normalization factors for each query.

```python
# data used for the analysis
# got with ./run_mix_vs_single.py --bench=MIXED --custom=ch_terminals=1,total_terminals=4,time=600,init=True
data_path =  "/home/alendit/master/runner/output/archive/configs 03.03.13 21:10/MIXED.raw"
```

```python
import numpy as np
import pandas as pd

def read_data(path):
    """Reads the data and computes a cumulative sum of the executed
    NEW_ORDER transactions"""

    s = pd.read_csv(path)
    
    s.columns = ['tt', 's', 'l']
           
    s['qcount'] = np.where(s['tt'] == 2, 1, 0)
    s['qcumsum'] = s['qcount'].cumsum()
    s['s'] -= s['s'].min()
    s['l'] /= 10. ** 6
    
    s.save("mixed.npy")
    
    return s
```

```python
data = read_data(data_path)
```

```python
import matplotlib.pyplot as plt

def plot_transactions(data):

    for tt in xrange(7, 29):
        fig = plt.figure()
        q = data[(data['tt'] == tt)]
        data.set_index('qcumsum')
        model = pd.ols(y=q['l'], x=q['qcumsum'])
        
        fig = plt.figure()
        q['l'].plot(ylim=(0, q['l'].max() * 1.1), marker='.')
        fig.savefig("tt_{}.svg".format(tt))
        model.y_predict.plot(color='red', title="Transaction {}".format(tt - 6))

```

```python
plot_transactions(data)
```

```python
def normalize(data):
    """Find the normalization factors for each query and returns
    a tuple of normalized data set with a list of normalization factors"""
    norm_data = []
    norm_factors = []
    for tt in xrange(7, 29):
        q = data[data['tt'] == tt]
        q.set_index('qcumsum')
        model = pd.ols(y=q['l'], x=q['qcumsum'])
        norm_factor = model.beta[0]
        norm_factors.append(norm_factor)
        q['l'] -= norm_factor * q['qcumsum']
        q = q.reset_index()
        norm_data.append(q)
    return norm_data, norm_factors
```

```python
norm_data, norm_factors = normalize(data)
```

```python
def plot_norm_data(norm_data):
    for count, query in enumerate(norm_data):
        figure()
        title("Query {}".format(count + 1))
        query.set_index('s')['l'].plot(ylim=(0, query['l'].max() * 1.1), marker='o')
```

```python
plot_norm_data(norm_data)
```
