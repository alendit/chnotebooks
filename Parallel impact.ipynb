Parallel workers impact
------------------------

This test examines the impact of running multiple parallel OLAP workers on the query throughput and latency.

Each workload consists of 17 phases corresponding to 0 to 16 OLAP workers. 

```python
from __future__ import division
# the data comes from
#data_path = "/home/alendit/master/runner/output/archive/multirun_config_2 11.03.13 14:12" # previous run
data_path = "/home/alendit/master/runner/output/archive/multirun_config_2 16.03.13 12:49/"
```

First the data is loaded and prepared. The raw data is a csv file with 5 columns - transaction type, start time in s since epoch, query latency in miliseconds, worker id (benchmark-specific) and phase number. We normalize start time so it shows seconds since test start and covert latency to seconds. Futher we add a column 'diff' which hasd a value of 1 at the end of each phase.

```python
from IPython.parallel import Client
from multiprocessing import Pool
import pandas as pd
import os

def read_data(path):
    """Used to distribute the reading task"""
    import pandas as pd
    
    data = pd.read_csv(path)
    data.columns = ['tt', 's', 'l', 'w', 'p']
    data['s'] -= data['s'].min()
    data['l'] /= 10. ** 6
    data['diff'] = data['p'].diff()
    return data
    
def prepare_data(path):
    """Loads the data into an array"""
    broker = Pool()
    #broker = Client().load_balanced_view()
    
    data = broker.map(read_data, 
        [os.path.join(path, "MIXED{}.raw".format(oltp_workers))\
                     for oltp_workers in xrange(1, 17)])

    #broker.wait()
    return data
```

```python
def compare_throughputs(run_data, title):
    """Plots two graphs of oltp and olap throughputs"""

    oltp = run_data[run_data['tt'] < 7]
    olap = run_data[run_data['tt'] >= 7]
    
    fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1)
    fig.suptitle(title)
    
    ax1.set_title("OLTP")
    ax1.set_xlabel("Time")
    ax1.set_ylim(0, 2500)
    ax2.set_title("OLAP")
    ax2.set_xlabel("Time")

    ax2.set_ylim(0, 3.8)
    
    for x in run_data[run_data['diff'] == 1]['s']:
        ax1.axvline(x, linestyle='--', color='cyan')
        ax2.axvline(x, linestyle='--', color='cyan')
    
    def plot_hist(ax, dataset, num_bins, color):
        bins, edges = np.histogram(dataset['s'], bins=num_bins)
        bins = bins.astype(float)
        bins /= np.diff(edges).astype(float)
        ax.plot(edges[:-1], bins, ls='steps', color=color)
    plot_hist(ax1, oltp, 600, 'blue')
    plot_hist(ax2, olap, 17, 'green')
    
    fig.tight_layout(pad=2)
    
    savefig("{}.svg".format(title))
```

```python
for count, run_data in enumerate(data):
    compare_throughputs(run_data, "{} OLTP workers".format(count + 1))
```

The graph analysis revealed that use of 5 workers offers an acceptable balance between OLTP and OLAP perfomance. We will use the run see the impact of additional OLAP queries. The latencies of the olap queries will be normalized using the normalization factors from previous tests (see Response times). 

```python
run_data = read_data(os.path.join(data_path, "MIXED4.raw"))
```

```python
def normalize(data):
    norm_data = []
    norm_factors = pd.load("norm_factors.npy")
    data['qcount'] = np.where(data['tt'] == 2, 1, 0)
    data['qcumsum'] = data['qcount'].cumsum()
    
    for tt in xrange(7, 29):
        q = data[data['tt'] == tt]
        q['l'] -= norm_factors[tt - 6] * q['qcumsum']
        # HACK: sometimes latency become negative, we just clip them to 0
        q['l'] = q['l'].clip(lower=0)
        norm_data.append(q)
    return norm_data
```

```python
def plot_olap_latencies(norm_data):
    for count, query in enumerate(norm_data):
        figure()
        for x in run_data[run_data['diff'] == 1]['s']:
            axvline(x, linestyle='--', color='cyan')
        title("Query " + str(count + 1))
        query = query.reset_index().set_index('s')
        query['l'].plot(ylim=(0, query['l'].max() * 1.1))
```

```python
norm_data = normalize(run_data)
plot_olap_latencies(norm_data)
```

As one can see, the query latency performance starts degrading in the phase 4 or 5 corresponding to 3 or 4 OLAP workers. The OLTP performance degrades with additional OLAP workers. We will test scenratios with 1 to 3 workers to determine the optimal configuration.

```python
import math

def get_metrics(data):
    #Transactional Throughput tpmCH
    tpmCH = len(data[data['tt'] == 2]) / (data['s'].max() - data['s'].min()) * 60
    # Geometric mean
    norm_data = normalize(data)
    gmean = 1
    for tt in norm_data:
        gmean *= tt['l'].mean()
    gmean = math.pow(gmean, 1 / len(norm_data))
    # duration per query set
    query_set_time = sum(tt['l'].mean() for tt in norm_data)
    # queries per hour
    QphCH = 60 * 60 / query_set_time * 22
    
    return tpmCH, gmean, query_set_time, QphCH
```
