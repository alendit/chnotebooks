Data analysis

```python
data_path =  "/home/alendit/master/runner/output/archive/multirun_config 13.01.13 16:41"
```

```python
import os
import numpy as np
from itertools import product
from multiprocessing import Pool

def analyse_throughput(location):
    import pandas as pd
    import numpy as np
    
    w, c = location
    
    df = pd.read_csv(os.path.join(data_path, "MIXED{}_{}.raw".format(w + 1, c + 1)), na_filter=False)
    df.columns = ['tt', 's', 'l']
    df = df[(df['tt'] >= 2) & (df['tt'] <=6)]
    return len(df) / (df['s'].max() - df['s'].min())

def analyse_metrics(location):
    import pandas as pd
    import numpy as np
    from evaluate import MetricsCalculator
    
    w, c = location
    
    mc = MetricsCalculator(os.path.join(data_path, "MIXED{}_{}.raw".format(w + 1, c + 1)), 2, keep_data=False)
    return mc

def analyse(analyse_method):
    pool = Pool(processes=4)
    throughput = pool.map(analyse_method, list(product(xrange(16), xrange(8))))
    
    throughput = np.array(throughput)
    throughput.shape = 16, 8

    return throughput
```

```python
throughput_data = analyse(analyse_throughput)
metrics_data = analyse(analyse_metrics)
```

Plotting the 3D graph
----------------------

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%pylab inline
%config InlineBackend.figure_format = 'svg'
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1, projection='3d')
xx, yy = np.meshgrid(range(1, 9), range(1, 17))
#zz = [r[x][y] for x, y in zip(xx, yy)]
surf = ax.plot_wireframe(xx, yy, throughput_data)
ax.set_xlabel("Analytical workers")
ax.set_ylabel("Trans workers")
ax.set_zlabel("Req/s")
ax.set_zlim(0, 2500)
fig.savefig("multirun_through.svg")
plt.show()
```
